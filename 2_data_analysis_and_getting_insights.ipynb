# My dataset

```python
df = pd.read_csv('Datasets/My_cleaned_dataset.csv',index_col='Unnamed: 0')
df.head()
```

```python
df.info()
```

```python
df['angiographic disease status'].unique()
```

## Based on `heart-disease.names` file I will make some replacements

```python
df['sex'].replace([1,0],['Male','Female'],inplace=True)
df['chest pain'].replace([1.0,2.0,3.0,4.0],
                        ['Typical Angina','atypical angina','non-anginal pain','asymptomatic'],inplace=True)
df['fasting blood sugar'].replace([1.0,0.0],
                        ['fps more than 120 mg/dl','fps less than 120 mg/dl'],inplace=True)
df['resting ECG results'].replace([0.0,1.0,2.0],
                        ['normal','abnormal','abnormal'],inplace=True)
df['exercise induced angina'].replace([1.0,0.0],
                        ['yes','no'],inplace=True)
df['slope'].replace([1.0,2.0,3.0],
                        ['upsloping','flat','down sloping'],inplace=True)
df['thalassemia'].replace([3.0, 6.0, 7.0],
                        ['normal','fixed defect','reversible defect'],inplace=True)
df['angiographic disease status'].replace([0,1],
                        ['no heart disease','heart disease'],inplace=True)
```

```python
df.head()
```

## Now I will save this new dataframe with a name "dataframe_for_analysis.csv"

```python
df.to_csv('Datasets/dataframe_for_analysis.csv')
```
