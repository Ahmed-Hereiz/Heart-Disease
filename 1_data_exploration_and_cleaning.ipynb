# Now it is time to replace '?' value by true values so :
## if angiographic disease status = 0
> - NaN values in thalassmia will be replaced by `3.0` <br><br>
> - NaN values in number of vessels will be replaced by `0.0`
## if angiographic disease status = 1
> - NaN values in thalassmia will be replaced by `7.0` <br><br>
> - NaN values in number of vessels will be replaced by `2.0`


### To do so I will create a new dataframe used *`to replace nan values only`* in the original dataframe

```python
heart_disease_data.replace('?',np.nan,inplace=True)

to_replace_nan = heart_disease_data[['angiographic disease status']]
to_replace_nan[['replace nan only in number of major vessels','replace nan only in thalassmia']] = 'empty till now'
```

```python
to_replace_nan.head()
```

```python
heart_disease_data['age'] = heart_disease_data['age'].astype('int')
heart_disease_data['sex'] = heart_disease_data['sex'].astype('object')
heart_disease_data['chest pain'] = heart_disease_data['chest pain'].astype('object')
heart_disease_data['fasting blood sugar'] = heart_disease_data['fasting blood sugar'].astype('object')
heart_disease_data['resting ECG results'] = heart_disease_data['resting ECG results'].astype('object')
heart_disease_data['exercise induced angina'] = heart_disease_data['exercise induced angina'].astype('object')
heart_disease_data['slope'] = heart_disease_data['slope'].astype('object')
heart_disease_data['angiographic disease status'] = heart_disease_data['angiographic disease status'].astype('object')
```

```python
for i in to_replace_nan.index:
    value = to_replace_nan['angiographic disease status'].loc[i]
    
    if value == 0:
        to_replace_nan['replace nan only in number of major vessels'].loc[i] = '0.0'
        to_replace_nan['replace nan only in thalassmia'].loc[i] = '3.0'
        
    elif value == 1:
        to_replace_nan['replace nan only in number of major vessels'].loc[i] = '2.0'
        to_replace_nan['replace nan only in thalassmia'].loc[i] = '7.0'
        
        
to_replace_nan.head()
```

```python
heart_disease_data['number of major vessels'] = heart_disease_data['number of major vessels'].fillna(to_replace_nan['replace nan only in number of major vessels'])
heart_disease_data['thalassemia'] = heart_disease_data['thalassemia'].fillna(to_replace_nan['replace nan only in thalassmia'])
```
